```python
import os

import matplotlib.pyplot as plt
import numpy as np
import math
import matplotlib.ticker as ticker

%matplotlib inline  
```

```python
# folder for output
output_dir="../output/"

# check if output folder exists. If not create it.
if not os.path.exists("../output/"):
    os.makedirs("../output/")
else:
# remove all files in output directory
# get all the files in the directory
    files=os.listdir(output_dir)
    #and remove them one by one
    for file in files:
        os.remove(output_dir + file)
```

```python
# color settings
#
# colors: https://matplotlib.org/3.1.0/gallery/color/named_colors.html
#
donker_blauw = '#0B1E74'
blauw        = '#203AB5'
geel         = 'yellow'
donker_geel  = 'gold'
wit          = 'white'
```

```python
nrows = 1
ncols = 1
index = 1

theta1 = 2 * np.linspace(-np.pi, np.pi, 600)
theta2 = theta1 + np.pi #rotate the plot a bit
r = 1 - np.cos(3*theta1)

plt.figure(figsize=(20,20),
           facecolor=wit)

plt.subplot(nrows,ncols,index, projection='polar')

# flower
plt.fill(theta1, r, facecolor=blauw, edgecolor=donker_blauw, linewidth=2)
plt.fill(theta2, r, facecolor=blauw, edgecolor=donker_blauw, linewidth=2)

# pistils
# =======
#
# https://matplotlib.org/3.1.0/api/_as_gen/matplotlib.pyplot.plot.html
#
theta_pistil = [np.pi*0, np.pi*2/3, np.pi*-2/3]
r_pistil     = [0.25   , 0.25     , 0.25]

plt.polar(theta_pistil,
          r_pistil,
          color=donker_geel,
          linewidth=0,
          marker='o',
          markeredgecolor=geel,
          markeredgewidth = 7,
          markersize=50)

#kruis - assen
plt.grid(b=True,axis='x',color=donker_blauw, linestyle='-', linewidth=10)

# set the circle tick texts
#
start_x, end_x = plt.xlim()
tick_spacing_x = (end_x-start_x)/6
xtickslbls_x = np.arange(start_x,end_x,tick_spacing_x)
plt.xticks(xtickslbls_x)
plt.tick_params(axis='x',direction='in', length=1, width=2, colors=donker_blauw, grid_color=donker_blauw, grid_alpha=0.5)
plt.tick_params(axis='y',direction='in', length=1, width=2, colors=donker_blauw, grid_color=donker_blauw, grid_alpha=0.5)

# first save the plot
plt.savefig(output_dir+"logo_ipheion.svg", format="svg")
plt.savefig(output_dir+"logo_ipheion_transparent.svg", format="svg", transparent=True)
plt.savefig(output_dir+"logo_ipheion.png", format="png")
plt.savefig(output_dir+"logo_ipheion_transparent.png", format="png", transparent=True)

# add a header text

# title setting
# =============
font = {'family': 'arial',
        'color':  donker_blauw,
        'weight': 'bold',
        'size': 60,
        }

plt.title('IPHEION logo', loc='center', fontdict=font)

# first save this plot also
plt.savefig(output_dir+"logo_ipheion_header.svg", format="svg")
plt.savefig(output_dir+"logo_ipheion_header.png", format="png")

# then show the plot
plt.show()
```
